```python
import numpy as np
import matplotlib.pyplot as plt

E = np.loadtxt('energies.txt')
T = np.loadtxt('temperatures.txt')

# Energy plots
fig,ax = plt.subplots(figsize=(9,7))
ax.plot(E[:,0],E[:,-1])
ax.set_xlabel('Iteration',fontsize=18)
ax.set_ylabel('e - E[-1] (eV)',fontsize=18) # Total energy - final energy
ax.tick_params(axis = 'both', labelsize = 13)
ax.grid()
ax.set_xlim([0,4000])

plt.savefig(f'Figures/Energy v iteration.pdf',
    bbox_inches = 'tight')

# Temperature plots
fig,ax = plt.subplots(figsize=(9,7))
ax.plot(T[:,0],T[:,-1])
ax.set_xlabel('Iteration',fontsize=18)
ax.set_ylabel('Temperature (K)',fontsize=18) # Total energy - final energy
ax.tick_params(axis = 'both', labelsize = 13)
ax.grid()
ax.set_xlim([0,4000])

plt.savefig(f'Figures/Temperature v iteration.pdf',
    bbox_inches = 'tight')
```
