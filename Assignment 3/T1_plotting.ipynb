```python
# Initialization
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True

# Importing the data
time,etot,epot,ekin,T = np.loadtxt('data/mdOutput.log',unpack=True,skiprows=1)
#time,etot,epot,ekin,T = np.loadtxt('Na-aimd/NaCluster24.log',unpack=True,skiprows=1)

# Fitting a curve to the energies
fit = np.polyfit(time,epot-epot[-1],4)
fit_func = np.poly1d(fit)
fit_x = np.linspace(time[0],time[-1],100)
fit_y = fit_func(fit_x)

# Energy plots
fig,ax = plt.subplots(figsize=(10,7))
ax.plot(time,epot-epot[-1],label='Data')
ax.plot(fit_x,fit_y,label='Fitted curve')
ax.set_xlabel(r'Time (ps)',fontsize=20)
ax.set_ylabel(r'$E_{pot}(t) - E_{pot}^{final}$ (eV)',fontsize=20) 
ax.tick_params(axis = 'both', labelsize = 15)
ax.grid()
#ax.set_xlim([0,2])
#ax.set_ylim([-.5,3.5])
ax.set_xlim([time[0],time[-1]])
ax.legend(fontsize=19)

plt.savefig(f'Figures/Temp Energy-time.pdf',
    bbox_inches = 'tight')

# Temperature plots
fig,ax = plt.subplots(figsize=(7,7))
ax.plot(time,T)
ax.set_xlabel(r'Time (ps)',fontsize=20)
ax.set_ylabel(r'Temperature (K)',fontsize=20) # Total energy - final energy
ax.tick_params(axis = 'both', labelsize = 15)
ax.grid()
ax.set_xlim([time[0],time[-1]])

plt.savefig(f'Figures/Temperature-time.pdf',
    bbox_inches = 'tight')
```
